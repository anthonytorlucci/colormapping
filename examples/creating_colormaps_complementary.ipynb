# creating colormaps - complementary
create colormaps from complementary html colors

```python
from context import colormappy

import matplotlib.pyplot as plt
import matplotlib.colors
import matplotlib.cm

from colormappy import ColorPointIndex, create_colormap
from colormappy_plotting import colormappy_figure, colormappy_attributes, scatter3D_Jpapbp, surface_sombrero
```

```python
c1, c2 = '#0080ff', '#ff8000'
c1_rgb = matplotlib.colors.hex2color(c1)
print(c1_rgb)
c2_rgb = matplotlib.colors.hex2color(c2)
print(c2_rgb)
```

```python
clr_point000 = ColorPointIndex(idx=0, c0=c1_rgb[0], c1=c1_rgb[1], c2=c1_rgb[2], cspace="sRGB1")
clr_point255 = ColorPointIndex(idx=255, c0=c2_rgb[0], c1=c2_rgb[1], c2=c2_rgb[2], cspace="sRGB1")
mappy_rgb = create_colormap(points=[clr_point000, clr_point255], interp_cspace="CAM02-UCS")
mappy_fig = colormappy_figure(mappy_rgb, f'complementary colors {c1} {c2} (CAM02-UCS)')
```
