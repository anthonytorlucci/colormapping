# create colormaps - metro ui
create a colormap from online color palette [digitalsynopsis](https://digitalsynopsis.com/design/beautiful-color-palettes-combinations-schemes/)

```python
from context import colormappy

import numpy
import matplotlib.pyplot as plt
import matplotlib.colors
import matplotlib.cm

from colormappy import ColorPointIndex, create_colormap
from colormappy_plotting import colormappy_figure, colormappy_attributes, scatter3D_Jpapbp, surface_sombrero
```

```python
# https://digitalsynopsis.com/design/beautiful-color-palettes-combinations-schemes/
mappy_name = 'Metro UI Colors' 
palette = [
    matplotlib.colors.hex2color('#d11141'),
    matplotlib.colors.hex2color('#00b159'),
    matplotlib.colors.hex2color('#00aedb'),
    matplotlib.colors.hex2color('#f37735'),
    matplotlib.colors.hex2color('#ffc425'),
]
mappy_fig = colormappy_figure(palette, mappy_name)
# mappy_fig.show()
```

```python
ind = numpy.linspace(start=0, stop=255, num=len(palette), endpoint=True, dtype=numpy.int16)
palpoints = []
for i in range(len(palette)):
    pal = palette[i]
    # print(pal)
    palpoints.append(ColorPointIndex(idx=ind[i], c0=pal[0], c1=pal[1], c2=pal[2], cspace="sRGB1"))


mappy_rgb = create_colormap(points=palpoints, interp_cspace='CAM02-UCS', interp_method='cubic')
mappy_fig = colormappy_figure(mappy_rgb, mappy_name)
# mappy_fig.show()
```

```python
fig = colormappy_attributes(mappy_rgb, mappy_name)
```

```python
fig = scatter3D_Jpapbp(mappy_rgb, mappy_name)
```

```python
fig = surface_sombrero(mappy_rgb, mappy_name)
```
