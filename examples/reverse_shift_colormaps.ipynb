# reverse and shift colormaps

```python
from context import colormappy

import matplotlib.pyplot as plt
import cmocean

from colormappy.colormappy import reverse_colormap, shift_colormap
from colormappy.utils import matplotlib_linsegcmap_to_rgba
from colormappy_plotting import colormappy_figure, colormappy_attributes, scatter3D_Jpapbp, surface_sombrero

mappy_rgb = matplotlib_linsegcmap_to_rgba(cmocean.cm.phase, 256, False)
mappy_name = 'cmocean phase'
fig = colormappy_figure(mappy_rgb, color_name=mappy_name)
```

```python
mappy_rev_rgb = reverse_colormap(rgb_arr=mappy_rgb)
mappy_name = 'cmocean phase reversed'
fig = colormappy_figure(mappy_rev_rgb, color_name=mappy_name)
```

```python
mappy_shf_rgb = shift_colormap(rgb_arr=mappy_rgb)
mappy_name = 'cmocean phase shifted'
fig = colormappy_figure(mappy_shf_rgb, color_name=mappy_name)
```
